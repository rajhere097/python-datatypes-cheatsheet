# Union 
Union combines elements from both sets (no duplicates)
set1 | set2

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result = set1 | set2
print(result)
```

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result = set1.union(set2) 
result # same way but using .union
```

# Intersection
Common elements between two sets
set1 & set2

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result1 =  set1 & set2
result1
```

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result1 =  set1.intersection(set2)
result1
```

# Difference
Elements in set1 that are NOT in set2
set1 - set2

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result2 = set1 - set2
result2
```

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result2 = set1.difference(set2)
result2 # elements in set1 that are not in set2
```

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result3 = set2.difference(set1)
result3 # elements in set2 which are not in set1
```

# Symmetric difference
Elements that are in either set1 or set2, but NOT in both
set1 ^ set2

```python
set1 = {12,12,14,15,16}
set2 = {15,16,17,18}
result2 = set1 ^ set2
result2
```

# R string
Raw string: A raw string means Python will not treat backslashes (\) as escape characters.
Normally, \ gives special meaning to characters after it, like \n (newline), \t (tab), \s, \d, etc.
But in a raw string, these escape sequences are not processed.

```python
cha = 'i am\good\at it'
cha
```

```python
cha = r'i am\good\at it'
cha
```

```python
# File path without raw string
path = 'C:\newfolder\data'
path # this will gives error because \n means newline
```

```python
# File path with raw string
path = r'C:\newfolder\data'
path # this will work as raw string keeps the backslashes as they are
```
